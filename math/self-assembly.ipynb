# Self-assembly

A recombination of evolutionary computing and cellular automata.

```python
import research as r
import scipy
import pandas

def fitness_plots(path, treatment):
    D = r.load_files(r.find_files(path, treatment+"_.*/fitness.dat"))

    figure()
    for i,g in D.groupby('trial'):
        plot(g['update'], g['max_fitness'])
    ylabel('Fitness')
    xlabel('Update')
    
    figure()
    r.quick_ciplot('update','max_fitness', D)
    ylabel('Fitness')
    xlabel('Update')
#    savefig("centroid.pdf")
    final = D[D['update']==D['update'].max()]
    print "\nDominant fitness:"
    print final.ix[final['max_fitness'].idxmax()]
    return D
```

```python
D = fitness_plots('../var/004-3d-fsm','ta0') # logic gates only
```

```python
D = fitness_plots('../var/004-3d-fsm','tb0') # logic+probabilistic gates only
```
